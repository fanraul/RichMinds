```python
dfm_saleshist_new = pd.DataFrame()

dfm_saleshist_new['SALES_TYPE'] = dfm_saleshist_old['Sales type']
dfm_saleshist_new['ORDER_NUMBER'] = dfm_saleshist_old['OrderNo'].astype('str') + '-' + dfm_saleshist_old['OrderLineNo'].astype('str')
dfm_saleshist_new['LOCATION_CODE'] = dfm_saleshist_old['Shipping location']
dfm_saleshist_new['SKU_CODE'] = ''
dfm_saleshist_new['SKU_CODE_ISCALA'] = dfm_saleshist_old['StockItemCode'].map(lambda x: str(x).strip())
dfm_saleshist_new['SKU_CODE_DESP'] = dfm_saleshist_old['StockItemDesc']

if processing == 'AKK':
    dfm_saleshist_new['SOLD_TO_ISCALA'] = dfm_saleshist_old['CompanyCode'] + dfm_saleshist_old['InvoiceCustomer Adjusted'].map(lambda x: str(x).strip())
    dfm_saleshist_new['SOLD_TO_DESP'] = dfm_saleshist_old['InvoiceCustomerName']
    dfm_saleshist_new['SHIP_TO_ISCALA'] = dfm_saleshist_old['CompanyCode'] + dfm_saleshist_old['OrderCustomer Adjusted'].map(lambda x: str(x).strip())
else:
    dfm_saleshist_new['SOLD_TO_ISCALA'] = dfm_saleshist_old['CompanyCode'] + dfm_saleshist_old['OrderCustomer'].map(lambda x: str(x).strip())
    dfm_saleshist_new['SOLD_TO_DESP'] = dfm_saleshist_old['OrderCustomerName']
    dfm_saleshist_new['SHIP_TO_ISCALA'] = dfm_saleshist_old['CompanyCode'] + dfm_saleshist_old['DeliveryCustomer'].map(lambda x: str(x).strip())

    
if 'SAP ship-to temp code' in dfm_saleshist_old.columns:
    dfm_saleshist_new['SHIP_TO_SAPTEMP'] = dfm_saleshist_old['SAP ship-to temp code'].map(lambda x: str(x).strip())
else:
    dfm_saleshist_new['SHIP_TO_SAPTEMP'] = 'nan'
    
dfm_saleshist_new['REQUESTED_DATE'] = dfm_saleshist_old['InvoiceDate']
dfm_saleshist_new['SHIP_DATE'] = dfm_saleshist_old['InvoiceDate']
dfm_saleshist_new['ETA_DATE'] = dfm_saleshist_old['InvoiceDate']
dfm_saleshist_new['ENTRY_DATE'] = dfm_saleshist_old['InvoiceDate']
dfm_saleshist_new['UOM_SCALA'] = dfm_saleshist_old['UOM']
dfm_saleshist_new['QTY_SCALA'] = dfm_saleshist_old['QuantityShipped']

dfm_saleshist_new['Customer Service'] = dfm_saleshist_old['Customer Service']
dfm_saleshist_new['Sales Rep'] = dfm_saleshist_old['Sales Rep']
dfm_saleshist_new['Sales Manager'] = dfm_saleshist_old['Sales Manager']
dfm_saleshist_new['Application 2'] = dfm_saleshist_old['MarketApp2']

#dfm_saleshist_new['COMPANY'] = dfm_saleshist_old['CompanyCode']

dfm_saleshist_new['TEST_MODE'] = 'FALSE'

dfm_saleshist_new['ORDER_NOTES'] =  dfm_saleshist_old['CompanyCode'].map(s_notes)
dfm_saleshist_new['WAVE'] =  'AW3-' + dfm_saleshist_new['ORDER_NOTES'].astype('str')

#mapping material
dfm_saleshist_new['SKU_CODE'] = dfm_saleshist_new['SKU_CODE_ISCALA'].map(s_material)


#get quantity conversion
# TODO: how to convert "250g Bot."?
def UoMconversion(s):
    if s['UOM_SAP'] == s['UOM_SCALA']:
        return s['QTY_SCALA']
    if s['UOM_SAP'] == 'KG' and s['UOM_SCALA'] == 'TO':
        return s['QTY_SCALA']*1000
    if s['UOM_SAP'] == 'TO' and s['UOM_SCALA'] == 'KG':
        return round(s['QTY_SCALA']/1000,2)
    if s['UOM_SAP'] == 'KG' and s['UOM_SCALA'] == '250g Bot.':
        return round(s['QTY_SCALA']*0.25,2)

dfm_saleshist_new['UOM_SAP'] = dfm_saleshist_new['SKU_CODE'].map(s_skuuom)

dfm_saleshist_new['QUANTITY'] = dfm_saleshist_new.apply(UoMconversion,axis = 1)

# use ix to reorder
dfm_saleshist_new  = dfm_saleshist_new.ix[:, ['SKU_CODE','ORDER_NUMBER','LOCATION_CODE','SOLD_TO',
               'SHIP_TO','REQUESTED_DATE','SHIP_DATE','ETA_DATE','ENTRY_DATE','QUANTITY','ORDER_NOTES',
               'SALES_TYPE','WAVE','TEST_MODE','SKU_CODE_ISCALA','SKU_CODE_DESP','SOLD_TO_ISCALA',
                'SOLD_TO_DESP','SHIP_TO_ISCALA','SHIP_TO_SAPTEMP','Customer Service','Sales Rep','Sales Manager','Application 2']]

#mapping customer
dfm_saleshist_new['SOLD_TO_SAP'] = dfm_saleshist_new['SOLD_TO_ISCALA'].map(s_customer)

def shiptoconvert(s):
    if s['SHIP_TO_SAPTEMP'] != np.nan and s['SHIP_TO_SAPTEMP'] != 'nan' and len(str(s['SHIP_TO_SAPTEMP']).strip() )> 0:
        return s['SHIP_TO_SAPTEMP']
    elif s['SHIP_TO_ISCALA'] in ('C8','H1','I1','K1','K2','CC','S1','J1','C4','T2'):
        return s['SOLD_TO_ISCALA']        
    else:
        return s['SHIP_TO_ISCALA']

dfm_saleshist_new['SHIP_TO_SAP'] = dfm_saleshist_new.apply(shiptoconvert,axis = 1 )
dfm_saleshist_new['SHIP_TO_SAP'] = dfm_saleshist_new['SHIP_TO_SAP'].map(s_customer)
#dfm_saleshist_new['SHIP_TO_SAP'] = dfm_saleshist_new['SHIP_TO_ISCALA'].map(s_customer)

dfm_saleshist_new2 = pd.merge(dfm_saleshist_new,dfm_PICASOcustomer, how = 'left', left_on = ['SOLD_TO_SAP','SHIP_TO_SAP'],
                              right_on = ['SoldTo ERP','ShipTo ERP'],sort  = False )

dfm_saleshist_new2['SOLD_TO'] = dfm_saleshist_new2['Sold-to code']
dfm_saleshist_new2['SHIP_TO'] = dfm_saleshist_new2['Ship-to code']

#dfm_saleshist_new2 = dfm_saleshist_new2[dfm_saleshist_new2.SALES_TYPE != '']

dfm_saleshist_new2.to_excel(output_excel_name)

print('processing successfully')
```
