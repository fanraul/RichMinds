```python
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
```

```python
#获得工厂清单
plant_dfm = pd.read_clipboard()
plant_dfm.tail()
```

```python
#获得SPT清单
SPT_dfm =pd.read_clipboard()
SPT_dfm.tail(100)
```

```python
#one to many merge plant list with MRP controller list
plantSPT_dfm = pd.merge(plant_dfm,SPT_dfm,on= 'Dummy' )
plantSPT_dfm.tail(200)
```

```python
#删除自己转到到自己的SPT
plantSPT_dfm2 = plantSPT_dfm[plantSPT_dfm['Plant'] != plantSPT_dfm['SPlant'] ]
plantSPT_dfm2.head(200)
```

```python
#准备删除已存在的plant+SPT组合
plantSPT_dfm2['plant_SPT'] = plantSPT_dfm2['Plant'] + '-' + plantSPT_dfm2['SPT']
plantSPT_dfm2.head()
```

```python
#获得已存在的工厂SPT清单
existplantSPT_dfm =pd.read_clipboard()
existplantSPT_dfm.tail()
```

```python
existplantSPT_dfm['plant_SPT'] = existplantSPT_dfm['Plant'] + '-' + existplantSPT_dfm['Special procurement']
existplantSPT_dfm.head()
```

```python
plantSPT_dfm3 = pd.merge(plantSPT_dfm2,existplantSPT_dfm, how = 'left', on = 'plant_SPT')
plantSPT_dfm3 = plantSPT_dfm3.fillna('new')
plantSPT_dfm4 = plantSPT_dfm3[plantSPT_dfm3['Special procurement'] == 'new']
plantSPT_dfm4.tail(200)
```

```python
plantSPT_dfm4['Special procurement'] = '7'
plantSPT_dfm4['Procurment type'] = 'F'
plantSPT_dfm4.tail(200)
```

```python
del plantSPT_dfm4['Plant_y']
del plantSPT_dfm4['plant_SPT']
plantSPT_dfm4.head()
```

```python
cols = ['Plant_x','SPT','Procurment type','Special procurement','SPlant','SPT text','Dummy']
# use ix to reorder
plantSPT_dfm4 = plantSPT_dfm4.ix[:, cols]
plantSPT_dfm4.tail(50)
```

```python
lang_dfm = pd.read_clipboard()
lang_dfm 
```

```python
plantSPTLang_dfm = pd.merge(plantSPT_dfm4,lang_dfm, how = 'inner', on = 'Dummy')
plantSPTLang_dfm.tail()
```

```python
del plantSPTLang_dfm['Dummy']
del plantSPTLang_dfm['SPlant']
del plantSPTLang_dfm['Special procurement']
del plantSPTLang_dfm['Procurment type']

```

```python
cols = ['Lang','Plant_x','SPT','SPT text']
# use ix to reorder
plantSPTLang_dfm = plantSPTLang_dfm.ix[:, cols]
plantSPTLang_dfm.head()
```

```python
writer = pd.ExcelWriter('newSPTplantlevel.xlsx')
plantSPT_dfm4.to_excel(writer,'Sheet1')
plantSPTLang_dfm.to_excel(writer,'Sheet2')
writer.save()
```
